```python
#how many items in one list are present in another using sets (time complexity of O(len(list1) * len(list2)))
def count_common_items(list1, list2):
    set2 = set(list2)
    return sum(1 for item in list1 if item in set2)

#how many items in one list are present in another using intersection (it is a bit more efficient --  time complexity of O(min(len(set1), len(set2))))
def count_common_items2(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    return len(set1 & set2)

#count number of common items in lists and calculate sorensen-dice index
def calculate_similarity(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    intersection = set1 & set2
    count = len(intersection)
    similarity = 2 * count / (len(set1) + len(set2))
    return count, similarity
```
